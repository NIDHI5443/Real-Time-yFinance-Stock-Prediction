```python
from pandas_datareader import data as pdr
import yfinance as yf

yf.pdr_override()

df = pdr.get_data_yahoo("%5EBSESN", start="2011-01-01", end="2021-08-03")
```

## Stock Graph of S&P BSE SENSEX

### Closing price v/s Year

```python
plt.figure(1,(10,5))
graph=plt.plot(df['Close'])
plt.xlabel("Year")
plt.ylabel("Closing price ($)")
plt.title("Stock Graph of S&P BSE SENSEX")
```

### Open Price v/s Year

```python
plt.figure(1,(10,5))
graph=plt.plot(df['Open'])
plt.xlabel("Year")
plt.ylabel("Open price ($)")
plt.title("Stock Graph of S&P BSE SENSEX")
```

### Adj Close price v/s Year

```python
plt.figure(1,(10,5))
graph=plt.plot(df['Adj Close'])
plt.xlabel("Year")
plt.ylabel("Adj Close price ($)")
plt.title("Stock Graph of S&P BSE SENSEX")
```

### Volume v/s Year

```python
plt.figure(1,(10,5))
graph=plt.plot(df['Volume'])
plt.xlabel("Year")
plt.ylabel("Volume")
plt.title("Stock Graph of S&P BSE SENSEX")
```
